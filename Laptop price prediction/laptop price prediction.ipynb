```python
def processor(str):
    if str =='Intel Core i5' or str =='Intel Core i7' or str =='Intel Core i3':
        return str
    else:
        if str.split()[0]=='Intel':
            return 'Other Intel Processor'
        else:
            return 'AMD Processor'

df['Cpus'] = df['Cpu_names'].apply(lambda x:processor(x))
```
